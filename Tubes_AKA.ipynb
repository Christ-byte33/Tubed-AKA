```python
import heapq
import matplotlib.pyplot as plt
from PIL import Image
import os

def dijkstra(graph, start, target):
    queue = [(0, start)]  # (total_weight, node)
    visited = set()
    distances = {node: float('inf') for node in graph}
    previous = {node: None for node in graph}

    distances[start] = 0

    while queue:
        current_weight, current_node = heapq.heappop(queue)

        if current_node in visited:
            continue

        visited.add(current_node)

        if current_node == target:
            break

        for neighbor, weight in graph[current_node]:
            if neighbor in visited:
                continue

            new_weight = current_weight + weight

            if new_weight < distances[neighbor]:
                distances[neighbor] = new_weight
                previous[neighbor] = current_node
                heapq.heappush(queue, (new_weight, neighbor))

    # Reconstruct the optimal path
    path = []
    current = target
    while current is not None:
        path.insert(0, current)
        current = previous[current]

    return path, distances[target]

# Graph representation for Naraka Bladepoint actions and loots
graph = {
    "Bronze": [("Silver", 2), ("Gold", 4), ("Loot Bronze", 1)],
    "Silver": [("Gold", 1), ("Platinum", 2), ("Loot Silver", 1)],
    "Gold": [("Platinum", 2), ("Diamond", 5), ("Loot Gold", 1)],
    "Platinum": [("Diamond", 3), ("Immortal", 4), ("Loot Platinum", 1)],
    "Diamond": [("Immortal", 2), ("Unrivaled Asura", 6), ("Loot Diamond", 1)],
    "Immortal": [("Unrivaled Asura", 1), ("Loot Immortal", 1)],
    "Unrivaled Asura": [],
    "Loot Bronze": [("Bronze", 1)],
    "Loot Silver": [("Silver", 1)],
    "Loot Gold": [("Gold", 1)],
    "Loot Platinum": [("Platinum", 1)],
    "Loot Diamond": [("Diamond", 1)],
    "Loot Immortal": [("Immortal", 1)]
}

# Hero recommendations based on rank
hero_recommendations = {
    "Bronze": [
        "Tarka Ji - Hero pemula yang solid dengan kemampuan seimbang.",
        "Viper Ning - Cocok untuk belajar mekanisme game dengan stun yang mudah digunakan.",
        "Matari - Hero support dengan mobilitas tinggi untuk membantu bertahan hidup."
    ],
    "Silver": [
        "Viper Ning - Stun dan crowd control yang cocok untuk pertempuran level menengah.",
        "Tarka Ji - Sangat baik untuk gaya bermain agresif.",
        "Matari - Mobilitas tinggi dengan kemampuan taktis untuk menyerang atau melarikan diri."
    ],
    "Gold": [
        "Matari - Bagus untuk pemain yang agresif dan untuk menangani rank yang lebih tinggi.",
        "Tianhai - Hero tank dengan kemampuan grappling yang sangat ofensif.",
        "Kurumi - Hero versatile yang bisa memberikan support healing atau damage."
    ],
    "Platinum": [
        "Matari - Mobilitas tinggi dan potensi damage besar untuk pemain yang sudah berpengalaman.",
        "Tianhai - Crowd control yang kuat dan damage burst tinggi.",
        "Kurumi - Hero fleksibel untuk mendukung permainan agresif atau bertahan."
    ],
    "Diamond": [
        "Tianhai - Hero tingkat ahli dengan taktik lanjutan dan kontrol medan perang.",
        "Valda Cui - Ideal untuk pemain yang mahir dengan kontrol area dan serangan AoE.",
        "Matari - Tetap andal untuk pemain berpengalaman yang mencari fleksibilitas dan mobilitas."
    ],
    "Immortal": [
        "Valda Cui - Cocok untuk pemain yang ingin memaksimalkan strategi dan AoE damage.",
        "Matari - Mobilitas kuat dengan strategi yang mematikan untuk tingkat lanjut.",
        "Tianhai - Hero cepat, cocok untuk pemain agresif yang ingin mendominasi."
    ],
    "Unrivaled Asura": [
        "Matari - Hero terbaik dengan performa luar biasa di semua situasi.",
        "Tianhai - Fleksibilitas luar biasa dan kemampuan yang kuat.",
        "Valda Cui - Pilihan solid untuk permainan tingkat atas."
    ]
}

# Input from the user
print("Daftar Rank: Bronze, Silver, Gold, Platinum, Diamond, Immortal, Unrivaled Asura")
start_rank = input("Masukkan rank awal Anda: ")

if start_rank not in graph:
    print("Rank tidak valid! Silakan masukkan rank yang benar.")
else:
    # Run Dijkstra's algorithm
    path, total_weight = dijkstra(graph, start_rank, "Unrivaled Asura")

    # Output results
    print("Jalur optimal untuk mencapai Unrivaled Asura:", " -> ".join(path))
    print("Total bobot (efisiensi atau risiko):", total_weight)

    # Display hero recommendations
    print("\nRekomendasi Hero untuk rank", start_rank + ":")
    for hero in hero_recommendations[start_rank]:
        print("- " + hero)

    # Display the map
    map_image_path = "nrPwHFTetBQj3zryituQsg-970-80.jpg"
    if not os.path.exists(map_image_path):
        print("Gambar peta tidak ditemukan, harap upload gambar terlebih dahulu.")
        from google.colab import files
        uploaded = files.upload()
        map_image_path = list(uploaded.keys())[0]

    if os.path.exists(map_image_path):
        img = Image.open(map_image_path)

        # Display map with annotations
        plt.figure(figsize=(12, 10), dpi=150)
        plt.imshow(img)
        plt.axis('off')
        plt.title("Peta Looting")

        # Add annotations to the map
        plt.figtext(0.5, 0.01, "- Warna merah menunjukkan area dengan musuh.\n- Area kuning menunjukkan zona dengan loot bagus.", wrap=True, horizontalalignment='center', fontsize=10)
        plt.show()
    else:
        print(f"Gambar peta tidak ditemukan pada lokasi: {map_image_path}")

import random

# Data Hero dan Kemampuan
heroes = {
    "Matari": {"skill": "Silent Flutter", "description": "Mobilitas tinggi untuk menyerang atau melarikan diri."},
    "Tarka Ji": {"skill": "Blackout", "description": "Serangan area besar, cocok untuk gaya agresif."},
    "Kurumi": {"skill": "Sacred Circle", "description": "Healing dan support untuk tim."},
    "Valda Cui": {"skill": "Nether Nightmare", "description": "Serangan AoE yang mematikan untuk kontrol area."},
    "Tianhai": {"skill": "Titan's Call", "description": "Crowd control dan damage burst tinggi."}
}

# Simulasi tips dan trik sebagai fungsi
def communicate_with_team():
    """Simulasi komunikasi dengan tim."""
    print("Strategi komunikasi: Gunakan ping untuk koordinasi dengan tim.")
    tips = [
        "Gunakan ping untuk menandai loot.",
        "Beritahu posisi musuh ke tim.",
        "Bagikan rencana menyerang atau bertahan melalui voice chat."
    ]
    for tip in tips:
        print(f"- {tip}")


def analyze_enemy_movement():
    """Analisis pergerakan musuh."""
    print("Analisis pergerakan musuh...")
    movements = ["Melarikan diri", "Menyerang langsung", "Menyergap", "Looting"]
    detected_movement = random.choice(movements)
    print(f"Musuh terdeteksi sedang: {detected_movement}")
    if detected_movement in ["Melarikan diri", "Looting"]:
        print("Serang segera saat mereka lengah!")
    elif detected_movement == "Menyergap":
        print("Hati-hati, gunakan grappling hook untuk keluar dari situasi bahaya.")
    else:
        print("Gunakan parry untuk bertahan, lalu balas serangan mereka.")


def late_game_strategy():
    """Strategi untuk zona akhir."""
    print("Strategi Zona Akhir:")
    print("- Pastikan armor dan senjata sudah diperbaiki.")
    print("- Tunggu momen yang tepat untuk menyerang musuh yang sibuk bertarung.")
    print("- Gunakan grappling hook untuk mengontrol medan perang.")
    print("- Jangan terlalu agresif, perhatikan posisi aman di zona kecil.")


def use_grappling_hook():
    """Tips penggunaan grappling hook."""
    print("Tips Menggunakan Grappling Hook:")
    print("- Gunakan untuk melarikan diri dari musuh.")
    print("- Kejar musuh yang mencoba kabur setelah menyerang.")
    print("- Grapple ke tempat tinggi untuk memantau area sekitar.")
    print("- Kombinasikan grappling hook dengan serangan jarak dekat untuk mengejutkan musuh.")


def manage_resources():
    """Manajemen sumber daya."""
    print("Manajemen Sumber Daya:")
    resources = ["Health Potion", "Armor Repair Kit", "Grappling Hook"]
    print("Pastikan Anda memiliki sumber daya berikut:")
    for resource in resources:
        print(f"- {resource}: Selalu cek jumlah yang tersisa.")

    critical_resources = random.choice(resources)
    print(f"Tips: Prioritaskan mengambil {critical_resources} di zona awal.")


def learn_parry_timing():
    """Latihan parry untuk pertahanan."""
    print("Latihan Timing Parry:")
    enemy_attack = ["Normal Attack", "Charged Attack", "Fake Attack"]
    for i in range(5):
        attack = random.choice(enemy_attack)
        print(f"Serangan musuh: {attack}")
        if attack == "Charged Attack":
            print("Tepat! Lakukan parry untuk membalas serangan.")
        else:
            print("Hati-hati! Tunggu charged attack untuk melakukan parry.")


def prioritize_loot():
    """Prioritas looting di awal game."""
    print("Prioritas Loot:")
    loot_items = ["Armor", "Weapon", "Soul Jade", "Health Potion", "Grappling Hook"]
    priorities = sorted(loot_items, key=lambda x: random.randint(1, 10), reverse=True)
    print("Urutan loot berdasarkan prioritas Anda:")
    for i, item in enumerate(priorities, 1):
        print(f"{i}. {item}")


def select_hero():
    """Pemilihan hero yang sesuai gaya bermain."""
    print("Pilih Hero:")
    for hero, info in heroes.items():
        print(f"- {hero}: {info['description']} (Skill: {info['skill']})")


def aim_training():
    """Latihan aim (hanya berupa panduan teks)."""
    print("Latihan Aim:")
    tips = [
        "Gunakan mode latihan untuk berlatih membidik target bergerak.",
        "Coba bidik kepala musuh untuk meningkatkan damage.",
        "Lakukan tracking musuh yang bergerak menggunakan grappling hook.",
        "Latih sensitivitas mouse Anda untuk mendapatkan kontrol yang lebih baik.",
        "Pastikan crosshair tetap sejajar dengan musuh saat berlari atau melompat."
    ]
    for tip in tips:
        print(f"- {tip}")


# Simulasi Tips
if __name__ == "__main__":
    print("\n[1] Komunikasi dengan Tim")
    communicate_with_team()

    print("\n[2] Analisis Pergerakan Musuh")
    analyze_enemy_movement()

    print("\n[3] Strategi Zona Akhir")
    late_game_strategy()

    print("\n[4] Penggunaan Grappling Hook")
    use_grappling_hook()

    print("\n[5] Manajemen Sumber Daya")
    manage_resources()

    print("\n[6] Latihan Timing Parry")
    learn_parry_timing()

    print("\n[7] Prioritas Loot")
    prioritize_loot()

    print("\n[8] Pemilihan Hero")
    select_hero()

    print("\n[9] Latihan Aim")
    aim_training()
```
